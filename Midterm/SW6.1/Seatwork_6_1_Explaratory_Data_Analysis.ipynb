```python
#Uploading  urinalysis_test.csv file and converting dataframe

filepath = '/content/urinalysis_tests.csv'
import numpy as np
import pandas as pd

data = pd.read_csv(filepath)
data
```

```python
#change the column name' Diagnosis'  to ' Lab Result'
df.rename(columns={"Diagnosis":"Lab Result"})
```

```python
#7. Create a new column Classification by diagnosis that display "UTI" if the value of outcome is 1 , otherwise "NO UTI"
df['Classification by diagnosis'] = df['Diagnosis'].apply(lambda x: 'UTI' if x =="POSITIVE" else 'No UTI')
df
```

```python
#8. Create a new dataframe "withUTI" that gathers data with UTI patients.
withUTI = df[df['Classification by diagnosis']=='UTI']
withUTI
```

```python
#9. Create a new dataframe "noUTI" thats gathers data with no UTI
noUTI = df[df['Classification']=='No UTI']
noUTI
```

```python
#10. Create a new dataframe "Minor" that gathers data with age 0 to 19
Minor = df[df['Age'] <= 19]
Minor
```

```python
#11. Create a new dataframe "Adult" that gathers data with age greater than 19
Adult = df[df['Age'] > 19]
Adult
```

```python
#12. Use numpy to get the average age , pH Level
#and specific gravity
# Extraction of data such as age,  ph level and  specific gravity
# in the dataset

ages= np.array(df['Age'])
pH_Level = np.array(df['pH'])
specific_gravity = np.array(df['Specific Gravity'])
```

```python
#Compute the following mean using numpy
ave_age = np.mean(ages)
ave_pH = np.mean(pH_Level)
ave_sg = np.mean(specific_gravity)
```

```python
#Display the following mean where age as a whole number and
# glucose, pH level and specific gravity  into  4 decimal value.

print(f"Average Age: {ave_age:.0f} years old")
print(f"Average pH Level: {ave_pH:.4f} ")
print(f"Average Specific Gravity: {ave_sg:.4f} ")
```

```python
#13.Use numpy to get the median age and pH Level, and specific gravity
# Using the extraction above we can get the median

#Compute the median of age and value using numpy
median_age = np.median(ages)
median_ph = np.median(pH_Level)
median_sg= np.median(specific_gravity)

#Display the median
print(f"Median age:{median_age} years old")
print(f"Median pH Level: {median_ph:.2f}")
print(f"Median specific gravity:{median_sg:.2f}")
```

```python
#15.Use numpy to get the standard deviation of the ph Level and Specific Gravity.


#compute for the std
std_ph = np.std(pH_Level, dtype=np.float32)
std_sg = np.std(specific_gravity,dtype =np.float32)

#Display the STD with 6 decimal places
print(f"Standard Deviation of pH Level: {std_ph:.6f}")
print(f"Standard Deviation of Specific Gravity: {std_sg:.6f}")
```
